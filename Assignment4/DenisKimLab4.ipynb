```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
SVMs = []
CEs = []

for i in range(1000):
    X = np.random.uniform(low=-10.0, high=10.0, size=(100,10))
    Y = np.random.randint(low=0, high=10, size=100)
    SVMs.append(calc_SVM())
    CEs.append(calc_CE())

    
plt.rcParams["figure.figsize"] = (20,10)
X0 = range(1000)
plt.plot(X0, SVMs, 'g', label = 'SVM')
plt.plot(X0, CEs, 'r', label = 'CE')

plt.xlabel('matrix #')
plt.ylabel('total loss')
plt.legend()
plt.show()
```

```python
def calc_SVM():
    SVM_loss = 0
    for i in range(100):
        L = 0
        for j in range(10):
            if j != Y[i]:
                L += max(0, (X[i][j] - X[i][Y[i]] + 1))
        SVM_loss += L
    return SVM_loss/100
```

```python
def calc_CE():
    CE_loss = 0
    one_hot_Y = np.zeros((100, 10))
    for i, y in enumerate(one_hot_Y):
        y[Y[i]] = 1
    
    for i in range(100):   
        CE_loss += np.dot(-1*one_hot_Y[i], np.log((X[i]+10)/np.sum(X[i]+10)))
    return CE_loss/100
```
