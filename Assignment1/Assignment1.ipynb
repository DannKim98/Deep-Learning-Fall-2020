```python
#Algorithm 1: Insertion Sort
def insertionSort(A):
    numOfComp = 0
    for i in range(1, len(A)):
        key = A[i]
        j = i - 1
        numOfComp += 1
        while j >= 0 and A[j] > key:
            A[j + 1] = A[j]
            j -= 1
            numOfComp += 1            
        A[j + 1] = key
    return numOfComp
```

```python
#Algorithm 2: Merge Sort

def merge(A, l, m, r):
    L = A.copy()[l:m + 1]
    R = A.copy()[m + 1:r + 1]
    i, j, k = 0, 0, l
    global numOfCompMerge
    
    while i < len(L) and j < len(R):
        if L[i] < R[j]:
            A[k] = L[i]
            i += 1
        else:
            A[k] = R[j]
            j = j + 1
        k = k + 1
        numOfCompMerge += 1
    
    # Add the rest of the arrays L and R to array A    
    while i < len(L): 
        A[k] = L[i] 
        i += 1
        k += 1
    while j < len(R): 
        A[k] = R[j] 
        j += 1
        k += 1

def mergeSort(A, l, r):
    if l < r:
        m = (l + r)//2
        mergeSort(A, l, m)
        mergeSort(A, m + 1, r)
        merge(A, l, m, r)
```

```python
import numpy as np

numsOfCompInsert = []
numsOfCompMerge = []
stepSize = 1000

for i in range(1, 11):
    arr1 = np.random.randint(1, 10000, i * stepSize)
    arr2 = arr1.copy()
    numsOfCompInsert.append(insertionSort(arr1))
    numOfCompMerge = 0
    mergeSort(arr2, 0, len(arr2) - 1)
    numsOfCompMerge.append(numOfCompMerge)
    print('Size ', i * stepSize, ' array, Insertion Sort: ', arr1)
    print('Size ', i * stepSize, ' array, Merge Sort: ', arr2)
print('\n---------------------------------------------------------\n')
print('Number of comparisons of insertion sort: ', numsOfCompInsert)
print('Number of comparisons of merge sort: ', numsOfCompMerge)
```

```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (9,6)
y1 = numsOfCompInsert
y2 = numsOfCompMerge
x = [i * stepSize for i in range(1, 11)]
plt.xlabel('number of elements in array')
plt.ylabel('number of comparisons')
plt.plot(x, y1, 'r', label = 'Insertion Sort')
plt.plot(x, y2, 'b', label = 'Merge Sort')
plt.legend()
plt.show()
```
